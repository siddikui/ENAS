```python
import numpy as np

x = np.array([4, 2, 8, 1, 6])

# First step: Obtain the indices that would sort x
sorted_indices_step1 = np.argsort(x)

# Second step: Obtain the ranks of the elements in x
rank_x = np.argsort(sorted_indices_step1)

print("Original array x:", x)
print("Indices that would sort x:", sorted_indices_step1)
print("Ranks of elements in x:", rank_x)
```

```python
import numpy as np

def spearman_rank_correlation(x, y):
    # Step 1: Rank the data
    rank_x = np.argsort(np.argsort(x))
    rank_y = np.argsort(np.argsort(y))

    # Step 2: Calculate the differences
    differences = rank_x - rank_y

    # Step 3: Square the differences
    squared_diff = differences ** 2

    # Step 4: Sum of squared differences
    sum_squared_diff = np.sum(squared_diff)

    # Step 5: Calculate Spearman's Rank Correlation Coefficient
    n = len(x)
    rho = 1 - (6 * sum_squared_diff) / (n * (n**2 - 1))

    return rho

# Example data
data_x = np.array([4, 2, 8, 1, 6])
data_y = np.array([8, 4, 16, 2, 12])

# Step 6: Calculate and print the correlation coefficient
correlation_coefficient = spearman_rank_correlation(data_x, data_y)
print(f"Spearman's Rank Correlation Coefficient: {correlation_coefficient}")
```
